```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.DataFrame({
    'Age' :[25,24,26,28,23,21,20],
    'Exp' : [5,6,7,8,4,5,4],
    'Salary' : [50000,75000,45000,65000,74000,45000,65000]
})
df
```

```python
df.describe()
```

```python
sc = StandardScaler()

scaled_df = pd.DataFrame(sc.fit_transform(df), columns=df.columns)
```
